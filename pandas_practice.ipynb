```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.hist(df["Age"].dropna())
plt.show()
```

```python
sns.histplot(df["Age"])
plt.show()
```

```python
sns.countplot(x="Sex", data=df)
plt.show()
```

```python
sns.barplot(
    x="Sex",
    y="Survived",
    data=df
)

plt.show()
```

```python
sns.boxplot(
    x="Pclass",
    y="Age",
    data=df
)

plt.show()
```

```python
sns.scatterplot(
    x="Age",
    y="Fare",
    data=df
)

plt.show()
```

```python
corr = df.corr(numeric_only=True)

sns.heatmap(
    corr,
    annot=True
)

plt.show()
```

```python
sns.pairplot(
    df[["Age","Fare","Pclass","Survived"]]
)

plt.show()
```

```python
sns.barplot(
    x="Pclass",
    y="Survived",
    hue="Sex",
    data=df
)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x="Pclass",
    data=df
)

plt.title("Passenger Class Distribution")
plt.xlabel("Passenger Class")
plt.ylabel("Count")

plt.show()
```
